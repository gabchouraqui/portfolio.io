# Problem Statement

```python
# Select only the numeric columns 
numeric_columns_prod = prod_df.select_dtypes(include=['int64', 'float64'])

# Define a threshold based on the Z score
z_score_threshold = 3

# Calculate the Z-scores for each numerical column
z_scores_prod = (numeric_columns_prod - numeric_columns_prod.mean()) / numeric_columns_prod.std()

# Identify outliers for each column based on the threshold
outliers_count_prod = (z_scores_prod > z_score_threshold) | (z_scores_prod < -z_score_threshold)

# Count the number of outliers per column
outliers_per_column_prod = outliers_count_prod.sum()
outliers_per_column_prod
```

```python
# Calculate the mean and standard deviation for each numerical column except for 'WellID', 'Lease', and 'API_UWI'
mean_values_prod = numeric_columns_prod.mean()
std_values_prod = numeric_columns_prod.std()

# Determine the lower and upper bounds for outliers
lower_bounds_prod = mean_values_prod - 2 * std_values_prod
upper_bounds_prod = mean_values_prod + 2 * std_values_prod

# Replace the outliers with the mean of each respective column
for index, row in prod_df.iterrows():
    for column in numeric_columns_prod.columns:
        if column not in ['WellID', 'Lease', 'API_UWI']:
            lower_bound = lower_bounds_prod[column]
            upper_bound = upper_bounds_prod[column]
            value = row[column]
            if (value < lower_bound).any() or (value > upper_bound).any():
                prod_df.at[index, column] = mean_values_prod[column]

# Outliers in the numerical columns (but for 'WellID', 'Lease', and 'API_UWI') have been replaced with the mean of the column
```

## 4 - Cost and Profitability Analysis

### STEP 1

The very first step will be to see the cost of each operator. To have the entire cost that might be difficult to find in such a dataset, a calculation with all the cost related to operator has been executed. 

The results show that EOG RESOURCES INC. has the higher cost and alsmot twice bigger than the other operators. 

```python
# Calculate cost per unit (USD per BOE) for each operator with drilling, completion, and tie-in costs
capital_efficiency_by_operator = metadata_df.groupby('ENVOperator').apply(lambda x: (x['DrillingCost_USDMM'] + x['CompletionCost_USDMM'] + x['TieInCost_USDMM']).sum())

# Calculate the cost per unit by operator
m_capital_efficiency_by_operator = capital_efficiency_by_operator.groupby('ENVOperator').mean().sort_values(ascending=False)

# Plotting capital efficiency by operator
plt.figure(figsize=(10, 6))
m_capital_efficiency_by_operator.plot(kind='bar', color='skyblue')
plt.xlabel('Operator', fontsize=12)
plt.ylabel('Total Cost', fontsize=12)
plt.title('Cost by Operator', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(fontsize=10)
plt.show()
```

### STEP 2

In this step, a column with the whole total cost has been added to the dataframe to in order to complete the next step and to have a global overview on the cost for each row.

```python
# Calculate total cost per well
metadata_df['TotalCost_USDMM'] = metadata_df['TotalWellCost_USDMM'] + metadata_df['DrillingCost_USDMM'] + metadata_df['CompletionCost_USDMM'] + metadata_df['TieInCost_USDMM']
metadata_df
```

### STEP 3

An intersting factor would be to calculate the profitability. 
In order to do that, the profitability has been calculated by dividing the cumulative production of BOE by the total cost (just added before).
And right after, a column with the rank of the profitability has been created to have a look on the most profiatbale contract.

```python
# Calculate profitability metrics
metadata_df['Profitability'] = metadata_df['CumProd_BOE'] / metadata_df['TotalCost_USDMM']

# Rank wells based on profitability
metadata_df['ProfitabilityRank'] = metadata_df['Profitability'].rank(ascending=False)

# Select top-ranked wells for portfolio optimization
optimized_portfolio = metadata_df.nlargest(10, 'ProfitabilityRank')
metadata_df
```
